```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


dfx = pd.read_csv("Bike Rental UCI dataset.csv")
dfx.head()
```

```python
def plotFeatures(col_list,title):
    plt.figure(figsize=(10, 14))
    i = 0
    print(len(col_list))
    for col in col_list:
        i+=1
        plt.subplot(7,2,i)
        plt.plot(dfx[col],dfx["demand"],marker='.',linestyle='none')
        plt.title(title % (col))   
        plt.tight_layout()
```

```python
col=["season","yr","mnth","hr","holiday","workingday","weathersit","temp","hum","windspeed","dayOfWeek","days"]
plotFeatures(col,"Relationship between %s and demand")
```

```python
class REG:
    
    def __init__(self,X,Y):
        self.X=X
        self.Y=Y
    
    def cost_function(self, omega):
        m = len(self.Y)
        J = np.sum((self.X.dot(omega) - self.Y) ** 2)/(2 * m)
        return J
    

    def gradient_descent(self, omega, alph):
    
        m = len(self.Y)
        iteration=0
        delta=1000
        CostL=[]
        omega_history=[]
        lastCostValue= self.cost_function(omega)+1
        
        while iteration<10000 and delta>0.00001:
            # Hypothesis Values
            h = self.X.dot(omega)
            # Difference omega/w Hypothesis and Actual Y
            loss = h - self.Y
            # Gradient Calculation
            gradient = self.X.T.dot(loss)/ m
            # Changing Values of omega using Gradient
            omega = omega - alph * gradient
            omega_history.append(omega)
            # New Cost Value
            cost = self.cost_function(omega)
            CostL.append(cost)
            
            delta = abs(lastCostValue - cost)
            lastCostValue = cost
            iteration+=1

        return omega, lastCostValue,CostL,h
```

```python
class relate:
    def __init__(self,h,Y):
        self.h=h
        self.Y=Y
         
    def rmse(self):
        mse = np.sum((self.h - self.Y)**2)
        rmse = np.sqrt(mse/m)
        return mse, rmse
    
    def r2(self):
        ssr = np.sum((self.h - self.Y)**2)
        sst = np.sum((self.Y - np.mean(self.Y))**2)
        r2_score = 1 - (ssr/sst)
        return ssr,sst,r2_score
```

```python
def Normalize(x):
    normal=(x-np.mean(x))/np.std(x)
    return normal
```
