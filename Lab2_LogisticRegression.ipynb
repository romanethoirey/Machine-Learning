# 1. Logistic Regression

```python
def gradientDescent(X,y,alpha=0.01,numIterations=10000):
    m=len(X.index)
    X_bias =np.ones((m,1)) #m is number of records in the dataset.
    X['bias']=X_bias
    xTrans = X.transpose()
    #w=np.random.randn()
    w=np.ones(len(X.columns))
    for i in range(numIterations):
        pred = sigmoid(np.dot(X, w))
        loss = pred - y
        gradient = np.dot(xTrans, loss)*(1/m)
        # update
        w = w - alpha * gradient
    return w
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def predict(X,y):
    w=gradientDescent(X,y)
    #print(w)
    y_pred=sigmoid(X.dot(w))
    #print(y_pred)
    classify=[1 if i >0.5 else 0 for i in y_pred]
    return classify
        
```

# 2. Features engineering

```python
import pandas as pd
import numpy as np
df=pd.read_csv('titanic.csv')
```

2.1 Sex

```python
df.Sex=df.Sex.replace('male',1)
df.Sex=df.Sex.replace('female',0)
df.Age=df.Age.fillna(df["Age"].mean())
```

2.2 Embarked

```python
df.Embarked.fillna('S')
df.Embarked=df.Embarked.replace('S',1)
df.Embarked=df.Embarked.replace('Q',2)
df.Embarked=df.Embarked.replace('C',3)
```

2.3 Civility

```python
df.loc[df.Name.str.contains("",na=False),'Civility']="3"
df.loc[df.Name.str.contains("Mr",na=False),'Civility']="1"
df.loc[df.Name.str.contains("Mrs",na=False),'Civility']="2"
df.loc[df.Name.str.contains("Miss",na=False),'Civility']="0"

df['Civility']=df['Civility'].astype(int)
```

2.4 Cabin

```python
df.loc[df.Cabin.str.contains("A",na=False),'Cabin']=0
df.loc[df.Cabin.str.contains("B",na=False),'Cabin']=1
df.loc[df.Cabin.str.contains("C",na=False),'Cabin']=2
df.loc[df.Cabin.str.contains("D",na=False),'Cabin']=3
df.loc[df.Cabin.str.contains("E",na=False),'Cabin']=4
df.loc[df.Cabin.str.contains("F",na=False),'Cabin']=5
df.loc[df.Cabin.str.contains("G",na=False),'Cabin']=6
df.loc[df.Cabin.str.contains("T",na=False),'Cabin']=7
df['Cabin']=df.Cabin.fillna(8)
```

2.5 isKid, isAlone

```python
df['isKid']=0
df.loc[df.Age<16,'isKid']=1
df['isAlone']=0
df.loc[(df.SibSp==0)&(df.Parch==0),'isAlone']=1
```

2.6 Age

```python
df['Age']=pd.qcut(df['Age'],5,labels=[1,2,3,4,5])
df['Age']=df['Age'].astype(int)
```

2.7 Fare

```python
df['Fare']=pd.qcut(df['Fare'],5,labels=[1,2,3,4,5])
df['Fare']=df['Fare'].astype(int)
```
